```python
def unir_arquivos(pri_letra):
    arquivos = pri_letra
    arquivo_saida = 'arquivo_unido.txt'
    with open(arquivo_saida, 'w') as saida:
        for nome_arquivo in arquivos:
            try:    
                with open(nome_arquivo, 'r') as entrada:
                    conteudo = entrada.read()
                    saida.write(conteudo)
                    saida.write('\n')
            except:
                print(f'Letra nao encontrada {pri_letra}')
    with open('arquivo_unido.txt','r', encoding='latin1') as f:
        t = f.read()
    return set(t.split())

def corrigir_texto(texto, matches):
    texto_corrigido = texto
    qt_erros = 0
    ajuste_rep = 0
    for erro in matches:
        i = erro.offset
        j = erro.errorLength 
        pri_letra = [f'ranking_palavras_{letra[0].lower()}.txt' for letra in erro.replacements]
        pri_letra = list(dict.fromkeys(pri_letra))
        palavras_ranking = unir_arquivos(pri_letra)
        rep = next((r.lower() for r in erro.replacements if r.lower() in palavras_ranking),None)
        if rep is None:
            try:                     
                rep = str(erro.replacements[0])  
            except:
                rep = texto_corrigido[i:j+i].upper()
        ajuste_rep = (ajuste_rep + len(rep) - j)*qt_erros
        texto_corrigido = texto_corrigido[:i+ajuste_rep] + rep + texto_corrigido[i+j:]
        qt_erros += 1
    return texto_corrigido

# Uso das funções

```
